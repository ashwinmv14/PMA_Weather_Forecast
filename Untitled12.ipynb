We can see the outliers from their corresponding boxplot, for features with numeric values

```python
def clip_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    df[column] = np.where(df[column] < lower_limit, lower_limit, df[column])
    df[column] = np.where(df[column] > upper_limit, upper_limit, df[column])
    return df
```

The function clip_outliers will,

- find what values are considered "outliers" using the IQR method.

- It does NOT remove rows from the dataset.

- Instead, it replaces extreme values (too low or too high) with the minimum or maximum acceptable value (based on IQR).

```python
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()

for col in numeric_cols:
    df = clip_outliers(df, col)
print(df.shape)

print("Outliers fixed successfully!")
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

print("Normalization completed!")
```

```python
datetime_cols = ['last_updated', 'sunrise', 'sunset', 'moonrise', 'moonset']
for col in datetime_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')

drop_cols = ['country', 'location_name', 'timezone', 'condition_text', 'wind_direction', 'moon_phase']
df.drop(columns=drop_cols, inplace=True)
```

A time series forecasting problem - predicting things like temperature or precipitation over time.

Columns like country, location_name, moon_phase are static or categorical. They don't change every hour or every minute like temperature or precipitation does.

They don't have a direct mathematical relationship with what you want to predict. Models like ARIMA, Exponential Smoothing, etc. require numerical, time-dependent input. Thus, keeping them will confuse the model without adding real value.

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_absolute_error, mean_squared_error

temperature = df['temperature_celsius']

temperature_daily = temperature.resample('D').mean()
temperature_daily = temperature_daily.dropna()

split_ratio = 0.8  # 80% training, 20% testing
split_index = int(len(temperature_daily) * split_ratio)
train = temperature_daily.iloc[:split_index]
test = temperature_daily.iloc[split_index:]
```

```python
model = ExponentialSmoothing(train, trend='add', seasonal=None)
fit_model = model.fit()

forecast = fit_model.forecast(steps=len(test))
```

```python
plt.figure(figsize=(12, 6))
plt.plot(train.index, train, label='Train', color='blue')
plt.plot(test.index, test, label='Actual Test', color='orange')
plt.plot(test.index, forecast, label='Forecast', color='green', linestyle='--', linewidth=3)  # Thicker dashed line
plt.title('Temperature Forecasting using Exponential Smoothing')
plt.xlabel('Time')
plt.ylabel('Temperature (Normalized)')
plt.legend()
plt.show()
```

```python
mae = mean_absolute_error(test, forecast)
rmse = np.sqrt(mean_squared_error(test, forecast))
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error

es_model = ExponentialSmoothing(train, trend='add', seasonal=None)
es_fit = es_model.fit()
es_forecast = es_fit.forecast(steps=len(test))

arima_model = ARIMA(train, order=(5, 1, 0))
arima_fit = arima_model.fit()
arima_forecast = arima_fit.forecast(steps=len(test))

def evaluate_model(true, pred, model_name):
    mae = mean_absolute_error(true, pred)
    rmse = np.sqrt(mean_squared_error(true, pred))
    print(f"{model_name} - MAE: {mae:.4f}, RMSE: {rmse:.4f}")

evaluate_model(test, es_forecast, "Exponential Smoothing")
evaluate_model(test, arima_forecast, "ARIMA")

ensemble_forecast = (es_forecast + arima_forecast) / 2

evaluate_model(test, ensemble_forecast, "Ensemble (Average of ES and ARIMA)")

plt.figure(figsize=(14, 7))
plt.plot(train.index, train, label='Train', color='blue')
plt.plot(test.index, test, label='Actual Test', color='orange')
plt.plot(test.index, es_forecast, label='Exponential Smoothing Forecast', linestyle='--', color='green')
plt.plot(test.index, arima_forecast, label='ARIMA Forecast', linestyle='--', color='purple')
plt.plot(test.index, ensemble_forecast, label='Ensemble Forecast', linestyle='--', color='red', linewidth=3)
plt.title('Temperature Forecasting: Comparing Models')
plt.xlabel('Time')
plt.ylabel('Temperature (Normalized)')
plt.legend()
plt.show()
```
